# Cross Validation Visualisations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from plotting_functions import visualise_cv

plt.rcParams['figure.figsize'] = (7, 3)
plt.rcParams['figure.dpi'] = 600
plt.rcParams['font.family'] = 'Times New Roman'
plt.rc('axes', labelsize=12)
plt.rc('legend', fontsize=10)
```

## BTC

```python
n_folds = 7
```

```python
coin = 'btc'
price_data = pd.read_parquet(f'../../2_data_processing/numeric_data/{coin}_price_data.parquet.gzip').iloc[:,0]
target_path = f'../../2_data_processing/numeric_data/{coin}_targets.parquet.gzip'
targets = pd.read_parquet(target_path)
```

### Visualise cross validation

```python
ax = visualise_cv(n_folds, targets.iloc[:,0], lw=12, skip_first=True)
x_min, x_max = ax.get_xlim()
new_ticks = np.linspace(x_min, x_max, 6)
ax.set_xticks(new_ticks)
ax.set_xticklabels(['0 %', '20 %', '40 %', '60 %', '80 %', '100 %'])
ax.set_xlabel('Proportion of available dataset')
plt.show()
```

```python
ax = visualise_cv(n_folds, targets.iloc[:,0], lw=12, skip_first=True)
x_min, x_max = ax.get_xlim()
ax.set_xticks([])
ax.set_xticklabels('')
ax.set_xlabel('Time', labelpad=7)
ax.set_ylabel('CV Fold', labelpad=7)
plt.show()
```

```python
ax = visualise_cv(n_folds, targets.iloc[:,0], lw=12, skip_first=True)
ax.set_title(f'Bitcoin Time Series Crossvalidation', fontsize=12);
```

```python
ax = visualise_cv(n_folds, targets.iloc[:,0], lw=12, skip_first=True, max_train_size=900)
ax.set_title('Bitcoin Time Series Crossvalidation (Constant Window)', fontsize=12);
```
